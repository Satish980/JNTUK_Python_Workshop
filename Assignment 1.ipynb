### 1. Create a class with below data members
- Product ID,
- Product name,
- price,
- Manufacturer name
    - a) provide necessary methods and override required magic methods
    - b) Create a List of above objects, sort based on price sort in reversing
    order, iterate and display

```python
#Create a class with data members Product ID,Product name, price,Manufacturer name
class Product:
    #constructor
    def __init__(self, Product_ID, Product_name, Price, Manufacturer_name):
        self.Product_ID=Product_ID
        self.Product_name=Product_name
        self.Price=Price
        self.Manufacturer_name=Manufacturer_name
    
    def __repr__(self):
        return "ID:"+str(self.Product_ID)+"  Product :"+str(self.Product_name)+"\tPrice:"+str(self.Price)+"\tManufacturer:"+str(self.Manufacturer_name)
    
    def __hash__(self):
        return hash((self.Product_ID))

    def __eq__(self, obj):
        return isinstance(obj, Product) and obj.Product_ID == self.Product_ID

    def __ne__(self, obj):
        return not self == obj

#A List of Product objects
print("************Product Details**************")
List = [Product(1,"Bike",40000,"Honda"),
        Product(2,"Cycle",4000,"Hero"),
        Product(3,"Car",4000000,"Audi"),
        Product(4,"Jeep",500000,"JEEP")]

List.append(Product(5,"BMW ",5000000,"BMW"))

#Printing the List of the Objects
print("\n************Product Details**************")
for obj in List:
    #printing the Object
    print(obj)
    
#sorting the List sort based on price sort in reversing order
List.sort(key=lambda x: x.Price, reverse=True)

#iterating and display the List of Object
print("\n************Product Details(Sorted)**************")
for obj in List:
    #printing the Object
    print(obj)
```

### 2. Use append(), and expand() slicing with positive/negative index, on above lists of objects. Use Set(), to remove duplicate objects, if any, and finally print elements of Set(with no duplicates). Consider only Product ID for duplication logic

```python
#appeding the List
List.append(Product(6,"Boat",50000,"Hyundai"))
#Printing the List of the Objects after appending
print("\n************Product After Appending Details**************")
for obj in List:
    #printing the Object
    print(obj)

#creating another List of Objects
List2=[Product(7,"Jet",1000000,"Airbus"),Product(3,"Truck",50000,"Tata Motors")]

#extending the List
List.extend(List2)

#Printing the List of the Objects after appending
print("\n************Extended Product Details**************")
for obj in List:
    #printing the Object
    print(obj)

#slicing with positive/negative index on above lists of objects
#positive index
List2= List[2:5];
#Printing the List of the Objects after appending
print("\n************Positive Index Product Details**************")
for obj in List2:
    #printing the Object
    print(obj)
    
#Negative index
List3=List[-5:-1]
print("\n************Negative Index Product Details**************")
for obj in List3:
    #printing the Object
    print(obj)

Object_Sets = set(List);

#Printing the List of the Objects after appending
print("\n************Set of Product Details**************")
for obj in Object_Sets:
    #printing the Object
    print(obj)
```
