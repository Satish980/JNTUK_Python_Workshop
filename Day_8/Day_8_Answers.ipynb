# 2.using MongoDB... Write a program to read all the reocords from MongoDB collection ( which was inserted in above question) and write the output to csv file

```python
myclient = pymongo.MongoClient("mongodb://localhost:27017")
mydb = myclient["rgukt"]
mycol = mydb["staff"]
fobj = open("jsontocsv.csv","w")
for i in mycol.find():
    street = i["street"]
    city = i["city"]
    zipc = i["zip"]
    state = i["state"]
    beds = i["beds"]
    baths = i["baths"]
    sqft = i["sq__ft"]
    typ = i["type"]
    sale_date = i["sale_date"]
    price = i["price"]
    latitude = i["latitude"]
    longitude = i["longitude"]
    record = ",".join([street,city,str(zipc),state,str(beds),str(baths),str(sqft),typ,sale_date,str(price),str(latitude),str(longitude)])
    fobj.write(record + "\n")
    print(record)
    
    
    
```
