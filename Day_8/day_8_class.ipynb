```python
#coverting json to csv
fobj = open("tocfsv.csv",'w')
with open('realestate.json') as f:
    data = json.load(f)
    for i in data:
        street = i["street"]
        city = i["city"]
        record = ",".join([street,city])
        fobj.write(record + "\n")
    
```
